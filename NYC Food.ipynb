```python
# This is a function

def fibonacci(n):
    a = b = 1
    for _ in range(n):
        b, a = a, b + a
    return b

# This is a test

def test_fibonacci():
    expected_sequence = [1, 1, 2, 3, 5, 8, 13]
    for n in range(7):
        actual = fibonacci(n)
        expected = expected_sequence[n]
        assert actual == expected, "Actual: " + \
                         str(actual) + \
                         " Expected: " + \
                         str(expected)
        
test_fibonacci()
        
for n in range(7):
    print n, fibonacci(n)
```

We're down to about 9000 entries! I want to allow the user to lookup via zipcode, so we should group by zipcode and sample now.

```python
from numpy.random import choice
def sample_zipcodes(data, count):                                                
    def sample(data, count):                                                       
        rows = choice(data.index.values, count)                                    
        sampled = data.ix[rows]                                                 
        return sampled                                                             
    sampled = {int(name): sample(d, count) for name, d in data}                  
    return sampled  

sampled = sample_zipcodes(camis_groups.groupby('ZIPCODE'), 10)
```

```python
#inspections.py
import pandas as pd                                                                
from numpy.random import choice                                                    
                                                                                   
                                                                                   
boros = {'manhattan': 1,                                                           
         'the bronx': 2,                                                               
         'brooklyn': 3,                                                            
         'queens': 4,                                                              
         'staten island': 5}                                                              
                                                                                   
                                                                                   
def _filter_latest_grades_(data):                                                  
    camis_groups = boro_df.groupby('CAMIS', as_index=False).first()
    camis_groups.dropna(subset=['CURRENTGRADE'], how='all')                           
    return camis_groups

                                                                                   
def _sample_zipcodes_(data, count):                                                
    def sample(data, count):                                                       
        rows = choice(data.index.values, count)                                    
        sampled = data.ix[rows]                                                    
        return sampled                                                             
    sampled = {int(name): sample(d, count).drop_duplicates() for name, d in data}                  
    return sampled                                                                 
                                                                                   
                                                                                   
def inspection_data(boro, samplesize):                                             
    webextract = 'data/WebExtract.txt'                                             
    data = pd.read_csv(webextract)                                                 
    key = boros[boro]                                                              
    boro_data = data.groupby('BORO').get_group(key)                                
    filtered = _filter_latest_grades_(boro_data)                                   
    zipcodes = filtered.groupby('ZIPCODE')                                         
    sampled = _sample_zipcodes_(zipcodes, samplesize)  
    return sampled 
```
